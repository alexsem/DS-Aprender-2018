```python
import pandas as pd
import numpy as np
import matplotlib as pyplot
```

```python
df = pd.read_csv('.\\Aprender2018-primaria-6.csv', delimiter=';', low_memory=False)
```

```python
df['mdesemp'].replace(' ', np.nan, inplace=True)
df.dropna(subset=['mdesemp'], inplace=True)

df['ldesemp'].replace(' ', np.nan, inplace=True)
df.dropna(subset=['ldesemp'], inplace=True)
```

```python
lpuntaje = pd.to_numeric(df["lpuntaje"].str.replace(',','.'), downcast='float')

df["lpuntaje"] = lpuntaje
```

```python
mpuntaje = pd.to_numeric(df["mpuntaje"].str.replace(',','.'), downcast='float')

df["mpuntaje"] = mpuntaje
```

```python
df.head()
```

```python
scatter_actividades = df[["lpuntaje","mpuntaje","ap13c"]].sample(500)
scatter_actividades['ap13c'] = pd.to_numeric(scatter_actividades['ap13c']).apply(lambda x: np.nan if x < 0 else x)

scatter_actividades.dropna(subset=['ap13c'], inplace=True)

axactividades = scatter_actividades.plot.scatter(x='lpuntaje',
                      y='mpuntaje',
                      c='ap13c',
                      colormap='viridis')

# ap13c: El alumno ayuda en casa cultivando, cosechando o cuidando animales
#1 Siempre
#2 Muchas veces
#3 Pocas veces
#4 Nunca

axactividades.set_title('Resultados segun si realiza actividades en casa')
axactividades.set_xlabel('puntaje en lengua')
axactividades.set_ylabel('puntaje en matematica')
```

```python
scatter_sector = df[["lpuntaje","mpuntaje","sector"]].sample(1000)
scatter_sector['sector'] = pd.to_numeric(scatter_sector['sector']).apply(lambda x: np.nan if x < 0 else x)

scatter_sector.dropna(subset=['sector'], inplace=True)

axsector = scatter_sector.plot.scatter(x='lpuntaje',
                      y='mpuntaje',
                      c='sector',
                      colormap='viridis')

# sector: Sector Publico o Privado
# 1 Publico
# 2 Privado

axsector.set_title('Resultados segun sector de la escuela')
axsector.set_xlabel('puntaje en lengua')
axsector.set_ylabel('puntaje en matematica')
```

```python
scatter_indice_socioeconomico = df[["lpuntaje","mpuntaje","isocioa"]].sample(1000)
scatter_indice_socioeconomico['isocioa'] = pd.to_numeric(scatter_indice_socioeconomico['isocioa']).apply(lambda x: np.nan if x < 0 else x)

scatter_indice_socioeconomico.dropna(subset=['isocioa'], inplace=True)

ax_indice_socioeconomico = scatter_indice_socioeconomico.plot.scatter(x='lpuntaje',
                      y='mpuntaje',
                      c='isocioa',
                      colormap='viridis')

# isocioa: Indice socioeconÃ³mico del alumno
# 1 Bajo
# 2 Medio
# 3 Alto

ax_indice_socioeconomico.set_title('Resultados en examenes segun indice socio-economico')
ax_indice_socioeconomico.set_xlabel('puntaje en lengua')
ax_indice_socioeconomico.set_ylabel('puntaje en matematica')
```

```python
scatter_caba = df[["lpuntaje","mpuntaje","cod_provincia"]].where(df["cod_provincia"] == 2).sample(500)
scatter_ba = df[["lpuntaje","mpuntaje","cod_provincia"]].where(df["cod_provincia"] == 6).sample(500)
scatter_salta = df[["lpuntaje","mpuntaje","cod_provincia"]].where(df["cod_provincia"] == 66).sample(500)
scatter_sanluis = df[["lpuntaje","mpuntaje","cod_provincia"]].where(df["cod_provincia"] == 74).sample(500)

ax_caba = scatter_caba.plot.scatter(x='lpuntaje', y='mpuntaje')

ax_caba.set_title('CABA')
ax_caba.set_xlabel('puntaje en lengua')
ax_caba.set_ylabel('puntaje en matematica')

ax_ba = scatter_ba.plot.scatter(x='lpuntaje', y='mpuntaje')

ax_ba.set_title('Buenos Aires')
ax_ba.set_xlabel('puntaje en lengua')
ax_ba.set_ylabel('puntaje en matematica')

ax_salta = scatter_salta.plot.scatter(x='lpuntaje', y='mpuntaje')

ax_salta.set_title('Salta')
ax_salta.set_xlabel('puntaje en lengua')
ax_salta.set_ylabel('puntaje en matematica')

ax_sanluis = scatter_sanluis.plot.scatter(x='lpuntaje', y='mpuntaje')

ax_sanluis.set_title('San Luis')
ax_sanluis.set_xlabel('puntaje en lengua')
ax_sanluis.set_ylabel('puntaje en matematica')
```
